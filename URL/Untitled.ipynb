```python
import pandas as pd
```

```python
df = pd.read_csv('unbalanced_train_segments.csv', sep =',', skiprows=2, names=['YTID','debut','fin','label1', 'label2', 'label3','label4'])
```

```python
df
```

```python
df = df.fillna('')
df = df.iloc[1:]
```

```python
df['debut'] = pd.to_numeric(df['debut'])
df['fin'] = pd.to_numeric(df['fin'])
```

```python
df['debut'] = df['debut'].astype(int)
df['fin'] = df['fin'].astype(int)
```

```python
def all_database(df):
    
   
    
   
        
    stream_lab1 = df.loc[df['label1'].str.contains('/m/0j6m2', regex=False)]   #stream1  
    stream_lab1.insert(7, "categorie", 'stream', allow_duplicates=False)
    stream_lab2 = df.loc[df['label2'].str.contains('/m/0j6m2', regex=False)]   #stream2 
    stream_lab2.insert(7, "categorie", 'stream', allow_duplicates=False)
    stream_lab3 = df.loc[df['label3'].str.contains('/m/0j6m2', regex=False)]   #stream3
    stream_lab3.insert(7, "categorie", 'stream', allow_duplicates=False)
    stream_lab4 = df.loc[df['label4'].str.contains('/m/0j6m2', regex=False)]   #stream4
    stream_lab4.insert(7, "categorie", 'stream', allow_duplicates=False)
    
    
    
    
    rain_lab1 = df.loc[df['label1'].str.contains('/m/06mb1', regex=False)]  #rain1
    rain_lab1.insert(7, "categorie", 'rain', allow_duplicates=False)
    rain_lab2 = df.loc[df['label2'].str.contains('/m/06mb1', regex=False)]  #rain2
    rain_lab2.insert(7, "categorie", 'rain', allow_duplicates=False)
    rain_lab3 = df.loc[df['label3'].str.contains('/m/06mb1', regex=False)]  #rain3
    rain_lab3.insert(7, "categorie", 'rain', allow_duplicates=False)
    rain_lab4 = df.loc[df['label4'].str.contains('/m/06mb1', regex=False)]  #rain4
    rain_lab4.insert(7, "categorie", 'rain', allow_duplicates=False)
    
    
    waterfall_lab1 = df.loc[df['label1'].str.contains('/m/0j2kx', regex=False)]  # Waterfall1
    waterfall_lab1.insert(7, "categorie", 'waterfall', allow_duplicates=False)
    waterfall_lab2 = df.loc[df['label2'].str.contains('/m/0j2kx', regex=False)]  # Waterfall2
    waterfall_lab2.insert(7, "categorie", 'waterfall', allow_duplicates=False)
    
    
    
    
    data_all = pd.concat([stream_lab1, stream_lab2, stream_lab3, stream_lab4, rain_lab1,
                         rain_lab2, rain_lab3, rain_lab4, waterfall_lab1, waterfall_lab2])
    data_all = data_all.reset_index(drop=True)
    

    
    return data_all
```

```python
all_data = all_database(df)
```
